```python
import pandas as pd
import re
import os
import pynmea2

f = open(r'C:\Users\Zedd\Desktop\gps_data2.txt',encoding = 'utf-8')
list1 = []
list2 = []
list3 = []
list4 = []
list5 = []
list6 = []
list7 = []
list8 = []
list9 = []
list10 = []
list11 = []
list12 = []
list13 = []
lines = f.readlines()
for line in lines:
    if line.startswith('$GPGGA'):   #查找GPRMC
        msg = pynmea2.parse(line)   #解析GPRMC，并存储
        list1.append(msg.lat)
        list2.append(msg.lat_dir)
        list3.append(msg.lon)
        list4.append(msg.lon_dir)
        list5.append(msg.gps_qual)
        list6.append(msg.num_sats)
        list7.append(msg.horizontal_dil)
        list8.append(msg.altitude)
        list9.append(msg.altitude_units)
        list10.append(msg.geo_sep)
        list11.append(msg.geo_sep_units)
        list12.append(msg.age_gps_data)
        list13.append(msg.ref_station_id)
dic = {"lat":list1,"lat_dir":list2,"lon":list3,"lon_dir":list4,"gps_qual":list5,"num_sats":list6,"horizontal_dil":list7,"altitude":list8,"altitude_units":list9,"geo_sep":list10,"geo_sep_units":list11,"age_gps_data":list12,"ref_station_id":list13}
df = pd.DataFrame(dic)
df.insert(0,'GPS_data','$GPGGA')
df.to_excel(r'C:\Users\Zedd\Desktop\gps_parse.xlsx')
f.close()
```
