```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体为黑体
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
```

```python
data1=pd.read_clipboard()
sns.scatterplot(y=data1['子身高'], x=data1['父身高'])
```

```python
data1.corr()
```

```python
sns.heatmap(data1.corr(),cmap='GnBu_r',annot=True,square=True)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lrmodel=LinearRegression()
x=data1[['父身高']]
y=data1[['子身高']]
lrmodel.fit(x,y)
```

```python
lrmodel.score(x,y)
```

```python
print("The linear model is: Y = {:.5} + {:.5}X".format(lrmodel.intercept_[0], lrmodel.coef_[0][0]))
```

```python
import statsmodels.api as sm
X = data1['父身高']
y = data1['子身高']
 
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```
