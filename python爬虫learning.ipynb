**<font size=3 color=blue>代码整合</font>**

```python
import requests
import json
import time

pic_id = []   #存储图片id，用于之后构造图片下载地址
urls = []
for page in range(0,1):
    url = 'https://unsplash.com/napi/photos?page={}&per_page=12'.format(page)
    header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.131 Safari/537.36"}
    req = requests.get(url, headers=header)
    jsn = req.text
    py = json.loads(jsn)   #要观察链接到的json文件是否存在错误字符
    for i in py:
        pic_id.append(i['id'])
        urls.append(i['urls']['small'])
    time.sleep(1)
for u in urls:   #循环读取图片下载地址并用直接写入文档的方式进行下载
    with open('C:/Users/Zedd/Desktop/images/'+pic_id[urls.index(u)]+'.jpg','ab+') as f:
        f.write(requests.get(u).content)
    print('保存第{}张图片成功'.format(urls.index(u)+1))
    time.sleep(1)
```

**<font size=4 color=blue>用定义函数的方式美化代码</font>**

```python
import requests
import json
import time

pic_id = []   #定义全局list存储信息
urls = []

def get_jsn(url):   #获取json文件 
    header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.131 Safari/537.36"}
    req = requests.get(url, headers=header)
    jsn = req.text
    return jsn

def get_pic_id(jsn):   #获取图片id
    py = json.loads(jsn)
    for i in py:
        pic_id.append(i['id'])
        
def get_download_url(jsn):   #获取图片下载地址
    py = json.loads(jsn)
    for i in py:
        urls.append(i['urls']['small'])
        
def download_pic():   #下载图片
    for u in range(0,24):
        with open('C:/Users/Zedd/Desktop/images2/'+pic_id[u]+'.jpg','ab+') as f:
            f.write(requests.get(urls[u]).content)
        print('保存第{}张图片成功'.format(u+1))
        time.sleep(1)
        
def main():   #主函数
    for page in range(0,2):
        url = 'https://unsplash.com/napi/photos?page={}&per_page=12'.format(page)
        jsnn = get_jsn(url)
        get_pic_id(jsnn)
        get_download_url(jsnn) 
        time.sleep(1)
    download_pic()
        
main()   #执行
    
```
