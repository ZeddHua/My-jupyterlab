```python
# 
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.image as image

# 
mpl.rcParams['font.sans-serif'] = ['SimHei']

# 
mpl.rcParams.update({'figure.autolayout': True})

# 
mpl.rcParams['axes.unicode_minus'] = False

# 
pd.set_option('display.float_format', lambda x: '%.2f' % x) 

# 
c = {'':'#00589F', '':'#003867', '':'#5D9BCF',
     '':'#999999', '':'#666666', '':'#CCCCCC',
     '':'#F68F00', '':'#A05D00', '':'#FBC171'}

# 
# filepath='./data/.xlsx'

#  Excel
df = pd.read_clipboard()

# 
x = df['']
y = df['']

# 
x2 = np.arange(len(x)+1) - 0.5
# 
y_cumsum = [0] + list(y.cumsum())
# 
y2 = y.cumsum()/y.sum()

# 
fig, ax = plt.subplots(figsize=(8, 6))

# 
ax.set_title('\n%.1f%%' % (y_cumsum[3]/y.sum()*100) + '23%\n', 
             fontsize=26, loc='left', color=c[''])

# 
ax.bar(x, y.sum(), width=1, color='w', edgecolor=c[''], zorder=0)

# 
ax.bar(x, y, width=1, color=c[''], edgecolor=c[''], zorder=1)

# 
ax.plot(x2, y_cumsum, ls='-', lw=2, color=c[''], label='', zorder=2)

# 
ax.hlines(y_cumsum[3], -0.5, 2.5, color=c[''], ls='--')
ax.vlines(2.5, 0, y_cumsum[3], color=c[''], ls='--')

# 
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# 
ax.spines['left'].set_position(('data', -0.51))

#  X 
ax.tick_params(axis='x', which='major', length=0)
ax.tick_params(axis='y', which='major', length=0)
ax.set_yticklabels([])

# 
ax.tick_params(labelsize=14, colors=c[''],labelrotation=70)

# 
for a, a2, b, b2, b3 in zip(x, x2[1:], y, y_cumsum[1:], y2):
    ax.text(a, b, '%.0f' % b, ha='center', va= 'bottom', fontsize=16, color=c[''])
    

#  Y 
ax.text(-1, y.sum(), '\n', fontsize=16, va='top', color=c[''])

# 
ax.text(1.5, y_cumsum[3]+10, '%.1f%% ' % (y_cumsum[3]/y.sum()*100), fontsize=16, color=c[''], va='bottom', ha='right', zorder=5)

plt.show()
```
