```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False

plt.style.use('seaborn-poster')

labels = ["", "","",""]
x = np.arange(4)
y =[1,1,0,0]
y1 =[4,0,0,0]
y2 =[1,1,1,1]
y3 =[0,0,1,1]
y4 =[5,1,0,0]
y5 =[1,1,1,1]
y6 =[1,1,2,2]
y7 =[8,4,0,4]
y8 =[4,4,2,2]
y9 =[1,1,1,1]

color=["#3FAA59","#B3DF72","#FEFFBE","#FDBB6C","#E54E35"]

plt.bar(x, y, align="center", tick_label=labels, label="")
plt.bar(x, y1, align="center", bottom=y, label="")
plt.bar(x, y2, align="center",  bottom=np.sum([y,y1],axis=0), label="")
plt.bar(x, y3, align="center",  bottom=np.sum([y,y1,y2],axis=0), label="")
plt.bar(x, y4, align="center",  bottom=np.sum([y,y1,y2,y3],axis=0), label="")
plt.bar(x, y5, align="center", bottom=np.sum([y,y1,y2,y3,y4],axis=0), label="")
plt.bar(x, y6, align="center", bottom=np.sum([y,y1,y2,y3,y4,y5],axis=0), label="")
plt.bar(x, y7, align="center", bottom=np.sum([y,y1,y2,y3,y4,y5,y6],axis=0), label="")
plt.bar(x, y8, align="center",bottom=np.sum([y,y1,y2,y3,y4,y5,y6,y7],axis=0), label="")
plt.bar(x, y9, align="center",  bottom=np.sum([y,y1,y2,y3,y4,y5,y6,y7,y8],axis=0), label="")

plt.xlabel("time period")
plt.ylabel("hr allocation")

#plt.legend(bbox_to_anchor=(1.10,1.12), ncol=4)
plt.legend(bbox_to_anchor=(1.35,1.0),borderaxespad = 0)

#plt.savefig(r'C:\Users\Zedd\Desktop\.jpg', dpi=300, bbox_inches='tight')

plt.show()

```
